## Brief analysis of  CO2 emissions with various bin widths
 Histograms were created to visualize the distribution of CO2 emissions. With different bin sizes (20 and 50 bins), we see how the distribution of emissions varies. The graphs show how most vehicles fall within a certain range of emissions.

From the histograms, it's likely we can conclude that the CO2 emissions for vehicles in this dataset vary, with the majority of vehicles falling within a mid-range for emissions.

```python
# Histogram for CO2 emissions with various bin widths
plt.figure(figsize=(10, 5))
sns.histplot(df['co2'], bins=20, kde=True, color='purple')
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (grams per mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('Distribution of CO2 Emissions')
plt.show()
```

```python
# Try with different bin sizes to compare
plt.figure(figsize=(10, 6))
sns.histplot(df['co2'], bins=50, kde=True, color='blue')
plt.title('Distribution of CO2 Emissions (50 Bins)')
plt.xlabel('CO2 Emissions (grams per mile)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('Distribution of CO2 Emissions (50 Bins)')
plt.show()
```

## Brief Analysis Of Fuel Efficiency
The city fuel efficiency histogram shows most vehicles cluster between 15 and 30 mpg, with a peak around 20-25 mpg. This suggests typical urban driving fuel consumption is moderate, reflecting stop-and-go traffic conditions. A small number of cars achieve higher mpg, likely hybrids or smaller models, but they’re less common. Overall, the data shows a typical spread with most vehicles falling into average city fuel efficiency ranges.

```python
# Histogram for 'city' fuel efficiency
plt.figure(figsize=(10, 8))
sns.histplot(df['city'], bins=30, kde=True, color='lime')
plt.title('Distribution of City Fuel Efficiency (mpg)')
plt.xlabel('City Fuel Efficiency (mpg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('city_fuel_efficiency.png')
plt.show()


```





## Brief Analysis of Engine Displacement

The histogram of engine displacement (`displ`) shows most vehicles have engines sized between 1.5L and 3.5L, with a noticeable peak around 2.0L to 3.0L. This indicates that mid-sized engines dominate the dataset, which makes sense for balancing power and fuel efficiency in everyday vehicles. There are fewer vehicles with very small (<1.5L) or very large (>4.0L) engines, showing that extreme engine sizes are less common in this sample.



```python
plt.figure(figsize=(8, 6))
sns.histplot(df['displ'], kde=True, color='green')
plt.title('Engine Displacement Distribution')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Frequency')
plt.savefig('engine_displacement.png')
plt.show()
```

## Brief Correlation Heatmap Analysis
The heatmap reveals some expected strong relationships. City, highway, and combined fuel efficiencies (city, highway, comb) are tightly positively correlated, which makes sense because they all measure MPG in different driving conditions.

CO2 emissions (co2) show strong negative correlations with fuel efficiency metrics—so as mpg goes up, emissions go down, which is logical. The feScore (fuel economy score) and ghgScore (greenhouse gas score) are positively correlated with mpg metrics and negatively correlated with CO2 emissions, indicating higher scores align with better fuel efficiency and lower emissions.

Overall, the heatmap confirms the interdependencies among fuel efficiency, emissions, and environmental impact scores in the dataset.

```python
# # Heatmap of correlations between numeric columns
cols = ['city', 'highway', 'comb', 'co2', 'feScore', 'ghgScore']
corr = df[cols].corr()
```

```python
# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Fuel, Facts, and Footprints", fontsize=14, pad=20)
plt.tight_layout()
# Save as PNG
plt.savefig("heatmap_fuel_footprints.png")
plt.show()
```
